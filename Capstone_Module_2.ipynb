#Data Understanding & Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
# Menampilkan Seluruh Data
df = pd.read_csv("/content/drive/MyDrive/Pakistan Largest Ecommerce Dataset.csv")
df
```

```python
# Melihat dimensi awal data
df.shape
```

```python
# Merubah Format kolom tanggal ke tipe data datetime
# Mengurutkan data berdasarkan tanggal dibuatnya transaksi
# Menambahkan kolom nama bulan
df['created_at'] = pd.to_datetime(df['created_at'])
df = df.sort_values(by = 'created_at')

df['Year'] = df['created_at'].dt.year
df['Month'] = df['created_at'].dt.month
df['Month_Name'] = df['created_at'].dt.month_name()
df.head()
```

## Missing Values

```python
# Melihat dan menyimpan nama kolom yang kosong
empty_col = []
for col in df:
  print(f"{col} = {df[col].nunique()}")
  if df[col].nunique() == 0:
    empty_col.append(col)
```

```python
# Menghapus kolom dengan data kosong
df_clean = df.drop(empty_col,axis = 1)
```

```python
# Menampilkan data setelah kolom kosong dihapus
df_clean
```

```python
# Menampilkan jumlah data kosong di tiap kolomnya
df_clean.isnull().sum()
```

```python
# Menghapus baris data dengan nilai kosong
df_clean = df_clean.dropna()
```

```python
# Menampilkan data setelah dihapus baris dengan nilai kosong
df_clean
```

```python
# Menampilkan jumlah  data kosong di tiap kolom setelah dibersihkan
df_clean.isnull().sum()
```

# Analisis

## Melihat Tren Transaksi Tiap Tahunnya

```python
# Membuat fungsi untuk mengelompokan data berdasarkan x dan y axis menurut inputan tahunnya
def df_filter_trend(year, xaxis, yaxis):
    df_year_filter=df_clean[df_clean['Year']== year]
    df_cat_sales= df_year_filter.groupby(xaxis)[yaxis].sum().reset_index()
    df_cat_sort= df_cat_sales.sort_values([xaxis], ascending=True)
    return df_cat_sort
```

```python
# Memanggil fungsi dengan parameter created at dan grand total untuk tiap tahunnya
trend_2016= df_filter_trend(2016, "created_at", "grand_total")
trend_2017= df_filter_trend(2017, "created_at", "grand_total")
trend_2018= df_filter_trend(2018, "created_at", "grand_total")
```

```python
#membuat plot dan subplot untuk tiap tahunnya
fig = make_subplots(rows=3, cols=1, subplot_titles=("Tren Jumlah Transaksi di tahun 2016", "Tren Jumlah Transaksi di tahun 2017",
                                                    "Tren Jumlah Transaksi di tahun 2018"))

fig.append_trace(go.Scatter(
    x=trend_2016["created_at"],
    y=trend_2016["grand_total"],
    name="Sales in 2016",
), row=1, col=1)

fig.append_trace(go.Scatter(
    x=trend_2017["created_at"],
    y=trend_2017["grand_total"],
    name="Sales in 2017",
), row=2, col=1)

fig.append_trace(go.Scatter(
    x=trend_2018["created_at"],
    y=trend_2018["grand_total"],
    name="Sales in 2018",
), row=3, col=1)


fig.update_layout(height=800, width=900, title_text="Tren Jumlah Transaksi tiap tahunnya")
fig.update_layout(
paper_bgcolor = " #f2f2f2",
plot_bgcolor = " #f2f2f2",
font=dict(family='sans-serif',
                  color='black',
                  size=10),


)
fig.show()
```

Analisis :     

Dapat dilihat bahwa berdasarkan data dari tahun 2016-2017 bahwa ada kenaikan signifikan pada bulan desember, dan berdasarkan tahun 2017-2018 ada kenaikan signifikan pada bulan Mei.

## Melihat Kategori barang paling banyak transaksi

```python
# Membuat fungsi untuk mengelompokan data berdasarkan x dan y axis menurut inputan tahunnya
def df_filter_cat(year, xaxis, yaxis):
    df_year_filter=df_clean[df_clean['Year']== year]
    df_cat_sales= df_year_filter.groupby(xaxis)[yaxis].sum().reset_index()
    df_cat_sort= df_cat_sales.sort_values([yaxis], ascending=False)
    return df_cat_sort
```

```python
# Memanggil fungsi dengan parameter category name 1 dan grand total untuk tiap tahunnya
cat_2016= df_filter_cat(2016, "category_name_1", "grand_total")
cat_2017= df_filter_cat(2017, "category_name_1", "grand_total")
cat_2018= df_filter_cat(2018, "category_name_1", "grand_total")
```

```python
#membuat plot dan subplot untuk tiap tahunnya
fig = make_subplots(rows=3, cols=1, subplot_titles=("Kategori dengan transaksi terbanyak di tahun 2016", "Kategori dengan transaksi terbanyak di tahun 2017",
                                                    "Kategori dengan transaksi terbanyak di tahun 2018"))

fig.append_trace(go.Bar(
    x=cat_2016["category_name_1"],
    y=cat_2016["grand_total"],
    name="2016"
), row=1, col=1)

fig.append_trace(go.Bar(
    x=cat_2017["category_name_1"],
    y=cat_2017["grand_total"],
    name="2017"
), row=2, col=1)

fig.append_trace(go.Bar(
    x=cat_2018["category_name_1"],
    y=cat_2018["grand_total"],
    name="2018"
), row=3, col=1)

fig.update_yaxes(title=dict(text= "Sales"))
fig.update_layout(height=800, width=900, title_text="Kategori dengan transaksi terbanyak di tiap tahun")
fig.update_layout(
paper_bgcolor = " #f2f2f2",
plot_bgcolor = " #f2f2f2",
font=dict(family='sans-serif',
                  color='black',
                  size=10),


)

fig.show()
```

Analisis : 

Tiap tahun kategori Mobiles & Tablets mendominasi penjualan.

## Melihat Status transaksi tiap tahunnya

```python
# Memanggil fungsi dengan parameter status dan grand total untuk tiap tahunnya
cat_2016= df_filter_cat(2016, "status", "grand_total")
cat_2017= df_filter_cat(2017, "status", "grand_total")
cat_2018= df_filter_cat(2018, "status", "grand_total")
```

```python
#membuat plot dan subplot untuk tiap tahunnya
colors = ['rgb(177, 127, 38)', 'rgb(205, 152, 36)', 'rgb(99, 79, 37)',
                     'rgb(129, 180, 179)', 'rgb(124, 103, 37)']


specs = [[{'type':'domain'}, {'type':'domain'}, {'type':'domain'}], [{'type':'domain'},{'type':'domain'}, {'type':'domain'}]]
fig = make_subplots(rows=2, cols=3, specs=specs, subplot_titles=['2016', '', '2017', '', 
                                                                 '2018'])

# Define pie charts
fig.add_trace(go.Pie(labels=cat_2016["status"], values=cat_2016["grand_total"], name='2016 Data',
                     marker_colors=colors), 1, 1)
fig.add_trace(go.Pie(labels=cat_2017["status"], values=cat_2017["grand_total"], name='2017 Data',
                     marker_colors=colors), 1, 3)
fig.add_trace(go.Pie(labels=cat_2018["status"], values=cat_2018["grand_total"], name='2018 Data',
                     marker_colors=colors), 2, 2)


# Tune layout and hover info
fig.update_traces(hoverinfo='label+percent+name+value', textinfo='label')
fig.update(layout_showlegend=False)

fig = go.Figure(fig)

fig.update_layout(title_text="Status Transaksi di tiap tahunnya", title_x=0.5),

fig.update_layout(
paper_bgcolor = " #f2f2f2",
plot_bgcolor = " #f2f2f2",
font=dict(family='sans-serif',
                  color='black',
                  size=12),


)
fig.show()
```

Analisis :    

Tiap tahunnya mayoritas transaksi masih dibatalkan walaupun tiap tahunnya jumlahnya berkurang.
